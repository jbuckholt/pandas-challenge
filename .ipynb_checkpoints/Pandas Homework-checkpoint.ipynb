```python
import pandas as pd 
# This will allow us to create file paths across operating systems
import os

# Module for reading CSV files
import csv
schools_csv = os.path.join('..', 'Pandas-challenge', 'schools.csv')
students_csv = os.path.join('..', 'Pandas-challenge', 'students.csv')

schools_df = pd.read_csv(schools_csv)
students_df = pd.read_csv(students_csv)

#schools_df.head()
#students_df.head()

# Combine the data into a single dataset.  
data_complete_df = pd.merge(students_df, schools_df, how="outer", on=["school_name", "school_name"])

#data_complete_df.head()
```

```python
total_schools = data_complete_df['school_name'].nunique()
total_students = data_complete_df['student_name'].nunique()
total_budget = schools_df['budget'].sum()
ave_math = data_complete_df['math_score'].mean()
ave_reading = data_complete_df['reading_score'].mean()

name = data_complete_df['school_name'].unique()
#passingmath_df = 
```

```python
#two new dataframes capturing the numbers of students passing math, reading
passmath_df = data_complete_df.loc[data_complete_df["math_score"] > 70, :]
passread_df = data_complete_df.loc[data_complete_df["reading_score"] > 70, :]

reading_pass = passread_df['reading_score'].count()
math_pass = passmath_df['math_score'].count()
# took the merged df and created a dataframe that filters down to only students that pass BOTH math and reading

passboth_df = data_complete_df[(data_complete_df["reading_score"] > 70) & (data_complete_df["math_score"] > 70)]

#count of students that passed both classes
passboth_class = passboth_df['student_name'].count()
print(passboth_class)
```

```python
#variables to hold % passing scores

per_mathpassing = (math_pass / total_students) * 100
per_readpassing = (reading_pass / total_students) * 100
per_passboth = (passboth_class / total_students) * 100
```

```python
result_df = pd.DataFrame([[total_schools, total_students, total_budget, ave_math, ave_reading, per_mathpassing, per_readpassing, per_passboth]], columns = ['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing'])
result_df
```

```python
#creating a new df for a different output
new_school_df = data_complete_df[["school_name", "type", "size", "budget", "math_score", "reading_score"]]
#new_school_df.head()
```

```python
# grouping by school name and getting the ave of math and reading columns
groupbyschool_df = new_school_df.groupby(["school_name"])
groupbyschool2_df = groupbyschool_df.mean()
#groupbyschool2_df.head()

```

```python
# using size of students and total budget to calculate budget per student

groupbyschool2_df["Budget_per_student"] = groupbyschool2_df["budget"] / groupbyschool2_df["size"]
#groupbyschool2_df.head()
```

```python
new_df = groupbyschool2_df[["size", "budget", "Budget_per_student", "math_score", "reading_score"]]
new_df.head()
# cleaning up column header names

new_df = new_df.rename(columns = {"size": "Total Students", 
                                   "budget": "Total Student Budget", 
                                   "Budget_per_student": "Per Student Budget",
                                   "math_score": "Average Math Score", 
                                   "reading_score": "Average Reading Score"})
```

```python
passing_math_df = new_school_df.loc[(new_school_df["math_score"] > 70),:]
#passing_math_df.()
groupbymath_df = passing_math_df.groupby(["school_name"])

groupbymath2_df = groupbymath_df.count()
groupbymath2_df = groupbymath2_df.rename(columns = {"math_score": "Total Passing Math"})
groupbymath2_df.head()
temp_df = pd.merge(new_df, groupbymath2_df, how="left", on=["school_name"])
temp_df = temp_df.drop(['budget', 'type', 'size', 'reading_score'], axis=1)
#temp_df.head()
```

```python
passing_reading_df = new_school_df.loc[(new_school_df["reading_score"] > 70),:]
#passing_math_df.()
groupbyread_df = passing_reading_df.groupby(["school_name"])

groupbyread2_df = groupbyread_df.count()
#school_type = school_sum_df['budget']
groupbyread2_df = groupbyread2_df.rename(columns = {"reading_score": "Total passing reading"})

#groupbyread2_df.head()
```

```python
temp2_df = pd.merge(temp_df, groupbyread2_df, how="left", on=["school_name"])
temp2_df = temp2_df.drop(['budget', 'type', 'size', 'math_score'], axis=1)
#temp2_df.head()
```

```python
passingboth_class_df = new_school_df.loc[(new_school_df["math_score"] > 70) & (new_school_df["reading_score"] > 70)]
#passing_math_df.()
passingbothgroup_df = passingboth_class_df.groupby(["school_name"])

passingbothgroup2_df = passingbothgroup_df.count()
passingbothgroup2_df = passingbothgroup2_df.rename(columns = {"size": "Passing Both Classes"})
#passingbothgroup2_df.head()
```

```python
temp3_df = pd.merge(temp2_df, passingbothgroup2_df, how="left", on=["school_name"])
temp3_df = temp3_df.drop(['budget', 'type', 'reading_score', 'math_score'], axis=1)
#temp3_df.head()
```

```python
temp3_df['% Passing Math'] = (temp3_df['Total Passing Math'] / temp3_df['Total Students']) * 100
temp3_df['% Passing Reading'] = (temp3_df['Total passing reading'] / temp3_df['Total Students']) * 100
temp3_df['% Passing Both'] = (temp3_df['Passing Both Classes'] / temp3_df['Total Students']) * 100
#temp3_df.head()
```

```python
temp3_df = temp3_df.drop(['Total Passing Math', 'Total passing reading', 'Passing Both Classes'], axis=1)
#temp3_df
```

```python
final_Best_df = temp3_df.sort_values(by='% Passing Both', ascending=False)
final_Best_df.head()
```

```python
final_worst_df = temp3_df.sort_values(by='% Passing Both', ascending=True)
final_worst_df.head()
```

```python
#defining new variables to consolidate scores per class per school
nine_df = students_df.loc[students_df['grade'] == "9th", ['grade', 'school_name', 'reading_score', 'math_score']]
ten_df = students_df.loc[students_df['grade'] == "10th", ['grade', 'school_name', 'reading_score', 'math_score']]
eleven_df = students_df.loc[students_df['grade'] == "11th", ['grade', 'school_name', 'reading_score', 'math_score']]
tweleve_df = students_df.loc[students_df['grade'] == "12th", ['grade', 'school_name', 'reading_score', 'math_score']]
```

```python
# grouping by school name and getting the ave of math and reading columns 9th grade
groupby9_df = nine_df.groupby(['school_name'])
groupby9_df = groupby9_df.mean()
groupby9_df = groupby9_df.rename(columns = {"reading_score": "9th Reading Ave", "math_score": "9th Math Ave"})
# grouping by school name and getting the ave of math and reading columns 10th grade
groupby10_df = ten_df.groupby(['school_name'])
groupby10_df = groupby10_df.mean()
groupby10_df = groupby10_df.rename(columns = {"reading_score": "10th Reading Ave", "math_score": "10th Math Ave"})
# grouping by school name and getting the ave of math and reading columns 11th grade
groupby11_df = eleven_df.groupby(['school_name'])
groupby11_df = groupby11_df.mean()
groupby11_df = groupby11_df.rename(columns = {"reading_score": "11th Reading Ave", "math_score": "11th Math Ave"})
# grouping by school name and getting the ave of math and reading columns 12th grade
groupby12_df = tweleve_df.groupby(['school_name'])
groupby12_df = groupby12_df.mean()
groupby12_df = groupby12_df.rename(columns = {"reading_score": "12th Reading Ave", "math_score": "12th Math Ave"})
```

```python
allscores_df = pd.merge(groupby9_df, groupby10_df, how="left", on=["school_name"])
allscores2_df = pd.merge(allscores_df, groupby11_df, how="left", on=["school_name"])
allscores3_df = pd.merge(allscores2_df, groupby12_df, how="left", on=["school_name"])
```

```python
allmath_df = allscores3_df.drop(['9th Reading Ave', '10th Reading Ave', '11th Reading Ave', '12th Reading Ave'], axis=1)
allmath_df
```
